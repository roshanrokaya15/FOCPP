```python
def check_for_coconut(string_list):
    """
    Returns True if 'coconut' is present in the list, 
    otherwise returns False.
    """
    for item in string_list:
        if item == "coconut":
            return True
    return False

# Demonstration code to show the function works
test_list_1 = ["apple", "banana", "coconut", "date"]
test_list_2 = ["cherry", "elderberry", "fig"]

print(f"Test 1 (Contains coconut): {check_for_coconut(test_list_1)}")
print(f"Test 2 (No coconut): {check_for_coconut(test_list_2)}")
```

```python
# Initialize an empty list to store inputs
user_strings = []

print("Enter strings one at a time (Press 'Enter' on an empty line to finish):")

while True:
    user_input = input("> ")
    
    # Check if the user pressed Enter without typing anything
    if user_input == "":
        break
    
    # Add the string to the list
    user_strings.append(user_input)

# Sort the list alphabetically
user_strings.sort()

# Display the strings in lower case
print("\nSorted strings in lower case:")
for s in user_strings:
    print(s.lower())
```
